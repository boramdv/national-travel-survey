## 관광·휴양 경험 유무

```python
# 관광·휴양 활동이 포함되지 않음, 1년 이하 해외여행 경험 column 제거
df = data.drop(['SA1_3', 'SA1_5', 'SA1_6'], axis=1)
```

```python
df.head(10)
```

```python
df['tour_num'] = df['SA1_1'] + df['SA1_2'] + df['SA1_4']
# df.head()
```

```python
# 여행 경험 없음 row 제거
tour_num_0 = df[df['tour_num'] == 0].index
len(tour_num_0)
```

```python
df = df.drop(tour_num_0)
len(df)
```

```python
# len(tour_num_0)+len(df)
```

## 최대 여행 횟수 이상 col & 1차 여행 무응답 row 삭제

```python
## 최대 여행 횟수 이상 조사한 column 삭제
df['tour_num'].max() # 5
```

```python
# tour_num <= 5이므로 여행6-8차 column 삭제
df = df.drop(df.columns[519:827], axis=1)  # 308개 column 제거
```

```python
df.shape # (15592, 2259)
```

```python
df.tail(20)
```

```python
## 1차 여행 무응답 row 삭제
# D_TRA1_CASE == NULL (87개) row 제거 
tra_1_0 = df[df['D_TRA1_CASE'] == '#NULL!'].index
df = df.drop(tra_1_0)
```

```python
df.shape # (15505, 2259)
```

```python
df.to_csv('2020_Domestic_Tour.csv', header=True, index=False)
```
